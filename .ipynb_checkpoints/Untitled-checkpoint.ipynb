```python
#looping control statements
# 3 type of statements
# break, continue pass
#BREAK STATMENT -->exit the loop immediately when encountered
#CONTINUE STATMENT--->skip the current iteration
#PASS STATEMENT--->its act as a place holder for future code reference
```

```python
#BREAK STATMENT -->exit the loop immediately when encountered
for i in range(1,10):
    if i==3:
        break
    print(i)
```

```python
#CONTINUE STATMENT--->skip the current iteration
for i in range(1,10):
    if i==3:
        continue
    print(i,end=" ")
```

```python
#PASS STATEMENT--->its act as a place holder for future code reference
for i in range(1,10):
    if i==3:
        pass
    print(i,end=" ")
```

```python
#reverse a number
num=int(input("Enter the number want to reverse"))
reverse=0
while num>0:
    toreverse = num % 10
    reverse=(reverse*10)+toreverse
    num //=10
    print(reverse)
```

```python
#scope of variables
#Three types of variables
#1. Local variable --->within the function
#2. Global variable--->Glabal keyword (outside of function),accessible from anywere in the file,
#3. Nonlocal--->used in nested unctions, refers to a variable in the nearest enclosing 
```

```python
#Local variable

def my_func():
    x=10 #Local variable
    print(x)

my_func()
```

```python
#Global variable
x=20
def my_func():
    global x 
    print(x)

my_func()
print(x)
```

```python
#Nonlocal
def outer():
    x="outer value"

    def inner():
        nonlocal x
        x="changed in inner"

    inner()
    print(x)

outer()
```

```python
#filter----> filter the records
#syntax --- > filter(function, iterable)

x=[23,4,5,17,89,90,67,34,256]

even=filter(lambda x:x%2==0,x)
list(even)
```

```python
names=["priya","arun","vishnu","sri","suba"]

filtered_names=filter(lambda a:a.startswith('s'),names)
list(filtered_names)
```

```python
#map-->vmodifier list, returns new list 
#syntax--> map(function, iterable)
names=["priya","arun","vishnu","sri","suba"]
changed_names=map(str.upper,names)
print(names)
print(list(changed_names))
```

```python
marks=[34,78,90,25,78,90,89,78,90,78,89]
updated_marks=map(lambda x:x+10,marks)
list(updated_marks)
```

```python
#reduce------> reduce list into single data
#syntax------> reduce(function, iterable)

from functools import reduce
total=[89,90,78,67,56]
final=reduce(lambda x,y:x+y,total)

final
```

```python
#question - filter strings which are length greater than 5
name1 = ["apple", "banana", "carrot", "Dessert", "fried", "strawberry"]
length_grearthan_five= filter(lambda x: len(x) > 5, name1)

list(length_grearthan_five)
```
