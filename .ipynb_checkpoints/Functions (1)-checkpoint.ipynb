```python
#function argument types

1.positional arguments
2.keyword arguments
3.default arguments
4,variable length keyword arguments
5.variable length positional arguments
```

```python
def logging(user_name,password):
    if user_name=="admin@123" and password=="admin":
        return True
    return False





user_name=input()
password=input()
is_logged_in=logging(user_name,password)
print(is_logged_in)
```

```python
#positional arguments

def student(name,age,course):
    print("Name",name)
    print("age",age)
    print("course",course)

student(24,"sri","ai")
```

```python
#default arguments

def data(name="guest"):
    print(f"welcome {name}")

data()
data("priya")
```

```python
#keyword arguments

def details(name,city,age,interest):
    print(name)
    print(city)
    print(age)
    print(interest)
details(name="vishnu",age=27,city="cbe",interest="cricket")
```

```python
#varaible length positional arguments

def total(*args):
    return sum(args)

total(90,89,78,67,56,56,45,44,34,12,34,56,7,89,9)
```

```python
#varaible length keyword arguments

def data(**kwargs):
    for key,value in kwargs.items():
        print(key,value)





data(name="priya",age=24,city="cbe",country="india",interest="dancing",hobby="cooking")
```

```python
# scope of varibles:
x=20
def my_func():
    x = 10  # local variable
    print(x)

my_func()
```

```python
x=20
def my_func():
    global x
    print(x)

my_func()
print(x)
```

```python
def outer():
    x = "outer value"

    def inner():
        nonlocal x
        x = "changed in inner"

    inner()
    print(x)

outer()
```
