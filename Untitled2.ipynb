# inheritance----------> one class inherites properties, methods from another class

# why inheritance?
code reusability

types of inheritance:
1.single inheritance
2.mutiple inheritance
3.multilevel inheritance
4.hierchical inheritance
5.hybrid unheritance

```python
#single inheritance: oru class la irukuratha innoru class la inheritate panrathud1

class animal:
    def show(self):
        print("Animals......")

class dog(animal):
    def sound(self):
        print("Dogs makes sound")

d1=dog()
d1.show()
d1.sound()
```

```python
class animal:
    def show(self):
        print("Animals......")

class dog(animal):
    def sound(self):
        print("Dogs makes sound")
    def show(self):
        super().show()   #super is a keyword can be used to get output of the show at the beginning
        print("Dogs show")
        animal.show(self)

d1=dog()
d1.show()
d1.sound()
```

```python
#multilevel inheritance

class grandparent:
    def work(self):
        print(f"Grandpa reading news paper")
class parent(grandparent):
    def display(self,name):
        print(f"parent name {name}")
    def work(self):
        print(f"parents working in a company")

class child(parent):
    def show(self):
        print(f"child...")
    def work(self):
        print(f"child studying")
        super().work()
c1=child()
c1.display("qwerty")
c1.work()
```

```python
#multiple

class father:
    def display(self,name,job):
        print(f"Name:{name}")
        print(f"Job:{job}")

class mother:
    def show(self,name,job):
        print(f"Name:{name}")
        print(f"Job:{job}")

class son(father,mother):
    def display(self,name,job):
        print(f"Name:{name}")
        print(f"Job:{job}")
        father.display(self,"ram","company")

s1=son()
s1.display("Naveen","AI Engineer")

s1.show("priya","House Wife")
```

```python
#hierchical inheritance

class parent:
    def show(self):
        print("Parents.....")
class child(parent):
    pass

class son(parent):
    pass
s1=son()
c1=child()

s1.show()
c1.show()
```

```python
class employee:
    def __init__(self,name,salary,job):
        self.name=name  #public data
        self.__salary=salary  #__ private data
        self._job=job #_ protected data
    def display(self,password):
        print(f"Employee name {self.name}")
        correct_password="admin@123"
        if password==correct_password:
            print(f"Salary {self.__salary}")
        else:
            print("Enter the correct password")
        print(f"job :{self._job}")

e1=employee("sripriya",30000,"fullstack devloper")
e2=employee("sri",200000,"AI Engineer")

password=input("Enter the password")
e1.display(password)
```

```python
import math

class shape:
    def area(self):
        pass

class rectangle(shape):
    def __init__(self,length,width):
        self.length=length
        self.width=width
    def area(self):
        return self.length*self.width

class circle(shape):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return math.pi*self.radius*self.radius

class triangle(shape):
    def __init__(self,b,h):
        self.b=b
        self.h=h

    def area(self):
        return 0.5*self.b*self.h

shape=[rectangle(10,2),circle(4),triangle(7,6)]

for shapes in shape:
    print(shapes.area())
```

```python
try:    #exception needs to be write like E caps - Exception
    a=10/0
    print(a)
except Exception as e:
    print(e)
```

```python
try:
    a=int(input("Enter the a number "))
    b=int(input("Enter the b number "))
    print(a+b)
except Exception as e:
    print(e)
except ValueError as e:
    print(e)
```

```python
try:
    a=int(input("Enter the a number "))
    b=int(input("Enter the b number "))
    print(a)
except ValueError:
    print("The value should be in integer")
```

```python
class agetoosmall(Exception):
    pass

def checkage(age):
    if age<0:
        raise agetoosmall("Age should be positive")
    else:
        print("valid age")


try:
    checkage(-9)
except Exception as e:
    print(e)
```

```python
class whichisbig(Exception):
    pass
    
def age (a,b):
    if a<b:
        raise whichisbig(a,b):
    else:
        print("Enter the valid age")
try:
    age(58,65)
except whichisbig as e:
    print(e)
    
```
