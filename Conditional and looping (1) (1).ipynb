#looping control statements

break, continue, pass

break-------->Exits the loop immediately when encountered.
continue------>skip the current iteration
pass------->it acts as place holder for future code reference

```python
for i in range(1,10):   #break
    if i==3:
        break 
    print(i)
```

```python
for i in range(1,10):   #continue
    if i==3:
        continue 
    print(i)
```

```python
for i in range(1,10):   #pass
    if i==3:
        pass
    print(i)
```
